```python
def shorten_name(title):
    title = str(title).lower()
    if 'sql' in title:
        return 'SQL'
    elif 'python' in title:
        return 'Python'
    elif 'excel' in title:
        return 'Excel'
    elif 'power bi' in title:
        return 'Power BI'
    elif 'machine learning' in title:
        return 'Machine Learning'
    elif 'photoshop' in title:
        return 'Photoshop'
    elif 'marketing' in title:
        return 'Marketing'
    elif 'finance' in title:
        return 'Finance'
    elif 'html' in title:
        return 'HTML'
    elif 'java' in title:
        return 'Java'
    elif 'ai' in title:
        return 'Artificial Intelligence'
    else:
        return 'Other'
    
df1['CourseNameShort'] = df1['CourseName'].apply(shorten_name)
```

```python
col = df1.pop('CourseNameShort')
df1.insert(df1.columns.get_loc('CourseName') + 1, 'CourseNameShort', col)
```

```python
df1['Category'].unique()
```

```python
category_map = {
    'Excel': 'Data Analysis',
    'Marketing': 'Marketing',
    'HTML': 'Programming',
    'SQL': 'Programming',
    'Java': 'Programming',
    'Finance': 'Finance',
    'Power BI': 'Data Analysis',
    'Python': 'Programming',
    'Photoshop': 'Design',
    'Artificial Intelligence': 'IT & Software'
}

df1['Category'] = df1['CourseNameShort'].map(category_map)
```

```python
df1['Rating'].value_counts()
```

```python
df1['RatingRounded'] = df1['Rating'].round()
```

```python
col1 = df1.pop('RatingRounded')
df1.insert(df1.columns.get_loc('Rating') + 1, 'RatingRounded', col1)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 15))
axes = axes.flatten()

# 1ï¸âƒ£ Top 10 Courses by Enrollments
top_course = df1.nlargest(10, 'Enrollments')[['CourseName', 'Enrollments']]
axes[0].barh(top_course['CourseName'], top_course['Enrollments'], color='skyblue')
axes[0].set_title('Top 10 Courses by Enrollments')
axes[0].invert_yaxis()

# 2ï¸âƒ£ Most Repeated Courses
course_counts = df1['CourseName'].value_counts().head(10).sort_values(ascending=True)
axes[1].barh(course_counts.index, course_counts.values, color='blue')
axes[1].set_title('Most Repeated Courses')

# 3ï¸âƒ£ Most Learned Skills
skill_counts = df1['CourseNameShort'].value_counts().sort_values(ascending=True)
axes[2].barh(skill_counts.index, skill_counts.values, color='purple')
axes[2].set_title('Most Learned Skills')

# 4ï¸âƒ£ Courses by Category
category_counts = df1['Category'].value_counts()
axes[3].bar(category_counts.index, category_counts.values, color='teal')
axes[3].set_title('Number of Courses by Category')
axes[3].tick_params(axis='x', rotation=45)

# 5ï¸âƒ£ Top 10 Best Rated Courses
avg_rating = df1.groupby('CourseName')['Rating'].mean().reset_index()
top_rated = avg_rating.nlargest(10, 'Rating')
axes[4].barh(top_rated['CourseName'], top_rated['Rating'], color='green')
axes[4].set_title('Top 10 Best Rated Courses')
axes[4].invert_yaxis()

# 6ï¸âƒ£ Best Rated Instructors
instructor_rating = df1.groupby('Instructor')['Rating'].mean().sort_values(ascending=False).head(10)
axes[5].bar(instructor_rating.index, instructor_rating.values, color='navy')
axes[5].set_title('Top 10 Best Rated Instructors')
axes[5].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

# 1ï¸âƒ£ Price vs Enrollments
sns.regplot(data=df1, x='Price', y='Enrollments', scatter_kws={'alpha':0.5}, ax=axes[0])
axes[0].set_title('Does Price Relate to Enrollments?')

# 2ï¸âƒ£ Price vs Rating
sns.regplot(data=df1, x='Price', y='Rating', scatter_kws={'alpha':0.5}, color='green', ax=axes[1])
axes[1].set_title('Does Price Affect Rating?')

# 3ï¸âƒ£ Enrollments vs Rating
sns.regplot(data=df1, x='Enrollments', y='Rating', scatter_kws={'alpha':0.5}, color='purple', ax=axes[2])
axes[2].set_title('Does More Enrollment Mean Better Rating?')

# 4ï¸âƒ£ Duration vs Rating
sns.regplot(data=df1, x='DurationHours', y='Rating', scatter_kws={'alpha':0.5}, color='red', ax=axes[3])
axes[3].set_title('Course Duration vs Rating')

plt.tight_layout()
plt.show()
```
